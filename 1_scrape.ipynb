# Notebook 1: Data Collection

This notebook fetches Google Play reviews for two mobile strategy games:
- **Clash of Clans** (`com.supercell.clashofclans`)
- **Rise of Kingdoms** (`com.lilithgame.roc.gp`)

Data is saved under `dataset/` directory as CSV files.  
These files will be used in later notebooks for embedding generation and competitive analysis.

**Requirements**: `google-play-scraper`, `pandas`, `time`, `os`.

```python
import pandas as pd
import os
import time
from google_play_scraper import reviews, Sort

# Common settings
OUTPUT_DIR = "dataset"
os.makedirs(OUTPUT_DIR, exist_ok=True)
TARGET = 100000          # Max reviews to fetch (actual limit ~10-20k)
PER_PAGE = 200
MAX_PAGES = 200
```

```python
# Clash of Clans
PACKAGE_NAME = "com.supercell.clashofclans"
OUTPUT_FILE = os.path.join(OUTPUT_DIR, "clashofclans.csv")

all_reviews = []
continuation_token = None
page = 0

print(f"Fetching {PACKAGE_NAME} ...")

while len(all_reviews) < TARGET and page < MAX_PAGES:
    try:
        limit = min(PER_PAGE, TARGET - len(all_reviews))
        result, continuation_token = reviews(
            PACKAGE_NAME,
            lang='en',
            country='us',
            sort=Sort.NEWEST,
            count=limit,
            continuation_token=continuation_token
        )
        if not result:
            break
        all_reviews.extend(result)
        page += 1
        print(f"Page {page}: {len(result)} reviews, total {len(all_reviews)}")
        if len(all_reviews) % 1000 == 0:
            pd.DataFrame(all_reviews).to_csv(OUTPUT_FILE, index=False, encoding='utf-8-sig')
        time.sleep(1)
    except Exception as e:
        print(f"Error: {e}, retry in 5s...")
        time.sleep(5)

pd.DataFrame(all_reviews).to_csv(OUTPUT_FILE, index=False, encoding='utf-8-sig')
print(f"Saved {len(all_reviews)} reviews to {OUTPUT_FILE}")
```
